b) Un diferencial de la flor se puede considerar como un $\Delta A$, dado por, 

$$
\Delta A = \frac{\Delta \theta}{2}\Big[\frac{r_i+r_{i+1}}{2}\Big]^2 
$$

Considerando ese diferencial de area, halle e imprima el area de la flor tomando 500 pasos para el 谩ngulo $\theta$ 

Pista si genera un `linspace`, ponga `retstep = True`



````python
flower_area = 0
## Tu codigo aqui
theta = np.linspace(0,2*np.pi,500)
r = 0.5*(np.cos(6*theta)+1)

def A(theta,r):
    Sum = 0 
    dtheta = theta[1] - theta[0]
    for i in range(len(theta)-1):
        dA = 0.5*dtheta*((r[i]+r[i+1])/2)**2
        Sum += dA
    return Sum
    
flower_area = A(theta,r)


## Codigo para verificar el resultado no modificar
flower_area
````

## Problema 2

Considere una esfera de radio $2$ , el diferencial de volumen en coordenadas esfericas est谩 dado por, 

$$
dV = \rho^2\sin{\phi} \,d\rho \,d\theta \,d\phi

El diferencial de volumen se puede simplificar para el caso de la esfera a, 

$$
\Delta V = \frac{\rho^3}{3}\sin{\phi} \,d\theta \,d\phi
$$

a) Calcule e imprima el volumen de una esfera de radio 2, con 500 pasos para cada coordenada.

````python
sphere_volume = 0
## Tu codigo aqui
theta = np.linspace(0,2*np.pi,500)
phi = np.linspace(0,np.pi,500)

def V(rho,theta,phi):
    Sum = 0
    dphi = phi[1] - phi[0]
    dtheta = theta[1] - theta[0]
    for j in range(len(phi)):
        for i in range(len(theta)):
            dV = (1/3)*(rho**3)*np.sin(phi[j])*dphi*dtheta
            Sum += dV
    return Sum
        
sphere_volume = V(2,theta,phi)
## Codigo para verificar el resultado no modificar
sphere_volume
````
